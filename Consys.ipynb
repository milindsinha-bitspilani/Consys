```
url = 'https://raw.githubusercontent.com/milindsinha-bitspilani/Consys/master/Midsem%20Test%20Marks%20(1).csv'
ms = pd.read_csv(url)
```

```
ms
```

```
grade=[]
```

```
for i in range(0,301):
  if ms['Marks (Out of 300)'][i]>206:
    grade.append(10)
  elif ms['Marks (Out of 300)'][i]<207 and ms['Marks (Out of 300)'][i]>199:
    grade.append(9)
  elif ms['Marks (Out of 300)'][i]<200 and ms['Marks (Out of 300)'][i]>166:
    grade.append(8)
  elif ms['Marks (Out of 300)'][i]<167 and ms['Marks (Out of 300)'][i]>150:
    grade.append(7)
  elif ms['Marks (Out of 300)'][i]<151 and ms['Marks (Out of 300)'][i]>108:
    grade.append(6)
  elif ms['Marks (Out of 300)'][i]<109 and ms['Marks (Out of 300)'][i]>90:
    grade.append(5)
  elif ms['Marks (Out of 300)'][i]<91 and ms['Marks (Out of 300)'][i]>59:
    grade.append(4)
  elif ms['Marks (Out of 300)'][i]<60 and ms['Marks (Out of 300)'][i]>32:
    grade.append(2)
  else:
    grade.append(0)
```

```
grade
```

```
ms['grade']=grade
```

```
ms
```

```
ms.dropna(inplace=True)
```

```
ms
```

```
plt.scatter(ms['rank'], ms['grade'],  color='blue')
plt.xlabel("Rank")
plt.ylabel("grade")
plt.show()
```

```
r_sq1 = model.score(x_, y1)
print('coefficient of determination:', r_sq1)
```
